```python
import pandas as pd
import os
import numpy as np
file = os.path.join("federal-firearm-licensees_Original.csv")
csv_file = pd.read_csv(file)
csv_file.count()
```

```python
csv_file = csv_file.drop(csv_file.index[0])
csv_file = csv_file.dropna(how="any")
csv_file.count()
```

```python
csv_file.head()
```

```python
license_count = csv_file["Premise City"].value_counts()
license_count.head()
```

```python
grouped_license_count = csv_file.groupby(['Premise State'])
grouped_city_license_count = pd.DataFrame(grouped_license_count["Premise City"].value_counts())
grouped_city_license_count.head()
```

```python
grouped_city_license_count.rename(columns = {"Premise State": "State", "Premise City": "City", "Premise City": "Licensees Count"}, inplace=True)
grouped_city_license_count.head(1000)
```
