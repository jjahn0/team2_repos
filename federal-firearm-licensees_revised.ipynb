```python
import pandas as pd
import os
import numpy as np
file = os.path.join("Resources" ,"federal-firearm-licensees_Original.csv")
csv_file = pd.read_csv(file)
csv_file.count()
```

```python
#Remove missing data
csv_file = csv_file.drop(csv_file.index[0])
csv_file = csv_file.dropna(how="any")
csv_file.count()
```

```python
#Testing
csv_file.head()
```

```python
#Count how many gun supplier per each State
license_count = csv_file.groupby("Premise State", as_index=False).count()
state = license_count["Premise State"]
count = license_count["Premise City"]
license_count.head()
```

```python
#Remove the rows with State that are not part of the 50 States
group_license_count["Gun Supplier Count"] = pd.DataFrame({"State": state,"Gun Supplier Count": count})
group_license_count = group_license_count[group_license_count.State != "DC"]
group_license_count = group_license_count[group_license_count.State != "GU"]
group_license_count = group_license_count[group_license_count.State != "MP"]
group_license_count = group_license_count[group_license_count.State != "PR"]
group_license_count = group_license_count[group_license_count.State != "VI"]
group_license_count = group_license_count.reset_index(drop=True)

group_license_count.head()
```

```python
#Reorder data frame and put State as 1st column
group_license_count = group_license_count[["State", "Gun Supplier Count"]]
group_license_count
```

```python
#Write data to CSV file
file_path = os.path.join("Resources","group_license_count.csv")
group_license_count.to_csv(file_path)
```
