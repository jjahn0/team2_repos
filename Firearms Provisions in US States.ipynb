```python
# Dependencies
import pandas as pd
import numpy as np
import os
```

```python
# Load CSV
gun_laws = os.path.join("raw_data.csv")
```

```python
# Import CSV
gun_laws_pd = pd.read_csv(gun_laws, encoding="iso-8859-1", low_memory=False)
gun_laws_pd.head()
```

```python
# Filter 2014 data
only_2014_gun_df = gun_laws_pd.loc[gun_laws_pd["year"] == 2014]
print(only_2014_gun_df.count())
only_2014_gun_df
```

```python
categories = list(only_2014_gun_df.columns.values)
```

```python
sum_list = [] 
for index in range(2,len(categories)-1):
    sum_list.append(only_2014_gun_df[categories[index]].sum())

sum_list
```

```python
del categories[135]
del categories[0]
del categories[0]
categories
```

```python
zip_it = dict(zip(categories, sum_list))
```

```python
gun_sum_pd = pd.DataFrame(columns=categories)
gun_sum_pd.append(zip_it, ignore_index=True)
```

```python
unsorted_file_path = os.path.join("Resources", "raw_data.csv")
sorted_file_path = os.path.join("Resources","gun_law_summary.csv")
gun_sum_pd.to_csv(sorted_file_path)
```
